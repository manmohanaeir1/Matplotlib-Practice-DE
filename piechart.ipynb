```python
import matplotlib.pyplot as plt

brands = ["OnePlus", "Apple", "Samsung", "Xaomi", "Nokia"]

x = [3,4,2,6,1]
c = ["red", "blue","lime", "orange", "pink"]
ex = [0,0,0,0,0.1 ]  # to highlight Nokia

plt.pie(x, labels = brands, colors = c, explode = ex, shadow = True, autopct = "%.2f", startangle = 180) # autopct for value in individual chart 
```
